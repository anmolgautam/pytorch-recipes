```python
import torch
```

```python
x = torch.tensor([[1], [2], [3]])
x.shape
```

```python
x.squeeze()
```

```python
y = torch.tensor([[[5]]])
y.shape
```

```python
y.squeeze()
```

```python
y.squeeze().shape
```

```python
z.unsqueeze(0)
```

```python
z.unsqueeze(0).shape
```
