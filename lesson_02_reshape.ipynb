# Lesson 02: Reshape

Goal: learn how tensor shape can be changed without changing the underlying values.

```python
import torch
```

```python
x = torch.tensor([1, 2, 3, 4, 5, 6])
x.shape
```

```python
x.reshape(2,3)
```

```python
x.reshape(2,3).shape
```

```python
x.reshape(3,2)
```

```python
x.reshape(2,2)
```

```python
x.numel()
```

```python
y = x.reshape(2,3)
y
```

```python
y.reshape(6)
```

```python
y.reshape(1,6)
```

```python
y.reshape(6,1)
```

```python
x.reshape(2,-1)
```

```python
x.reshape(-1,2)
```

```python
x.reshape(-1,-1)
```

```python
z = torch.tensor([[1, 2, 3], [4, 5, 6]])
z.shape
```

```python
z.reshape(3,2)
```

```python
z.reshape(1,6)
```
